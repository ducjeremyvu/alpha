### Write from File

```python
symbol = "US500"

# file outside of project,
# contains minute data with columns: Time, Open, High, Low, Close, Volume
#
#

file_path_csv = "/Users/ducjeremyvu/trading/price_data/US500_Minute_20251107_1228.csv"

insert_query = f"""
INSERT INTO ohlcv_minute (symbol, time, open, high, low, close, volume)
SELECT
    '{symbol}' AS symbol,  -- << your asset name here
    (CAST(Time AS TIMESTAMP) AT TIME ZONE 'UTC') AS time,
    CAST(Open AS DOUBLE) AS open,
    CAST(High AS DOUBLE) AS high,
    CAST(Low AS DOUBLE) AS low,
    CAST(Close AS DOUBLE) AS close,
    CAST(Volume AS BIGINT) AS volume
FROM read_csv_auto('{file_path_csv}')
ON CONFLICT (symbol, time) DO NOTHING;
"""



print(insert_query)


con.execute(insert_query)
```

```python
# delete

con.execute("DELETE FROM ohlcv_minute WHERE symbol = 'US500';")
```

## Queries

```python
con.execute(f"""
select * from ohlcv_minute
where symbol = 'US500'
order by time desc            
limit 10;
""").df()
```

```python
con.execute("""
            
SELECT 
    symbol,     
    timezone('UTC',time) as utc_time, 
    open,
    high,
    low,
    close,
    volume 
FROM ohlcv_minute 
where strftime(utc_time, '%Y-%m-%d') == '2025-11-05' 
order by utc_time            
""").df()
```
